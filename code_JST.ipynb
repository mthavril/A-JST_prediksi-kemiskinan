```python
# Import library yang dibutuhkan
import pandas as pd
from google.colab import files
from sklearn.model_selection import train_test_split

# Upload dataset
uploaded = files.upload()

# Membaca dataset (gantilah 'data_miskin.xlsx' dengan nama file sebenarnya)
df = pd.read_excel('data_miskin.xlsx')

# Menampilkan beberapa baris pertama dataset sebelum pembersihan
print("Data sebelum pembersihan:")
print(df.head())

# Pembersihan dataset
# Contoh: Menghapus baris dengan nilai-nilai yang hilang
df_cleaned = df.dropna()

# Menampilkan seluruh dataset setelah pembersihan
print("\nData setelah pembersihan:")
print(df_cleaned)



```

```python
# Fungsi normalisasi sigmoid sesuai rumus yang diberikan
def custom_sigmoid_normalization(x, min_val, max_val):
    return 0.8 * (x - min_val) / (max_val - min_val) + 0.1
df_cleaned = pd.DataFrame(df_cleaned)

# Normalisasi sigmoid pada kolom numerik sesuai rumus yang diberikan
numerical_columns = df_cleaned.columns[2:]  # Mulai dari kolom ke-2
df_normalized = df_cleaned.copy()

for column in numerical_columns:
    df_normalized[column] = custom_sigmoid_normalization(df_cleaned[column], df_cleaned[column].min(), df_cleaned[column].max())

# Menampilkan beberapa baris dari dataset yang telah dinormalisasi
print("\nData setelah normalisasi:")
print(df_normalized)

```

```python
# Mengasumsikan df_normalized adalah DataFrame yang telah dinormalisasi
# Memisahkan data untuk pelatihan (2012-2020) dan pengujian (2021-2022)
data_pelatihan = df_normalized[df_normalized.columns[2:-2]][0:]
target_pelatihan = df_normalized[df_normalized.columns[-2]][0:]

data_pengujian = df_normalized[df_normalized.columns[2:-1]][0:]
target_pengujian = df_normalized[df_normalized.columns[-1]][0:]

# Menampilkan set pelatihan
print("Data Pelatihan:")
print(data_pelatihan)
print("\nTarget Pelatihan:")
print(target_pelatihan)

# Menampilkan set pengujian
print("\nData Pengujian:")
print(data_pengujian)
print("\nTarget Pengujian:")
print(target_pengujian)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split

# Assuming df_normalized is your normalized DataFrame
# Selecting features and target column
features = df_normalized[df_normalized.columns[2:-2]]
target = df_normalized[df_normalized.columns[-2]]

# Splitting data into training and testing sets
train_data, test_data, train_target, test_target = train_test_split(features, target, test_size=0.2, random_state=42)

# Creating a simple neural network model
model = models.Sequential()
model.add(layers.Dense(9, activation='relu', input_shape=(train_data.shape[1],)))
model.add(layers.Dense(92, activation='relu'))
model.add(layers.Dense(1, activation='linear'))

# Compiling the model
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mae'])

# Training the model using training data
history = model.fit(train_data, train_target, epochs=100, batch_size=16, validation_split=0.2)

# Evaluating the model on testing data
test_loss, test_mae = model.evaluate(test_data, test_target)

# Getting predictions from the model on testing data
predictions = model.predict(test_data)

print(f'Test Loss: {test_loss}, Test MAE: {test_mae}')

# Menampilkan prediksi untuk semua data uji
all_predictions = model.predict(features)
print("\nPrediksi untuk semua data uji:")
for i, prediction in enumerate(all_predictions):
    print(prediction[0])
```
